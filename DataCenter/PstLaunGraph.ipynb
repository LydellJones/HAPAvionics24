```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfSRD = pd.read_csv('sRADF621cl.csv')
dfEZ = pd.read_csv('2024-06-15-serial-3373-flight-0011.csv')
```

```python
# Plot for the first DataFrame (dfSRD)
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('Speed and Acceleration', color=color)
ax1.plot(dfSRD['Time'], dfSRD['Speed'], label='Speed', color='blue')
ax1.plot(dfSRD['Time'], dfSRD[' IMUaccelY'], label='Acceleration', color='green')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis
color = 'tab:orange'
ax2.set_ylabel('Altitude', color=color)  # We already handled the x-label with ax1
ax2.plot(dfSRD['Time'], dfSRD[' BMPalt'], label='Altitude', color='orange')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # Otherwise the right y-label is slightly clipped
fig.suptitle('Speed, Acceleration, and Altitude over Time (SRAD)', y=1.02)
fig.legend(loc="upper right", bbox_to_anchor=(0.1, 1))

plt.show()

# Plot for the second DataFrame (dfEZ)
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('Speed and Acceleration', color=color)
ax1.plot(dfEZ['time'], dfEZ['speed'], label='Speed', color='blue')
ax1.plot(dfEZ['time'], dfEZ['acceleration'], label='Acceleration', color='green')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis
color = 'tab:orange'
ax2.set_ylabel('Altitude', color=color)  # We already handled the x-label with ax1
ax2.plot(dfEZ['time'], dfEZ['altitude'], label='Altitude', color='orange')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # Otherwise the right y-label is slightly clipped
fig.suptitle('Speed, Acceleration, and Altitude over Time (EzMini)', y=1.02)
fig.legend(loc="upper right", bbox_to_anchor=(0.1, 1))

plt.show()
```
